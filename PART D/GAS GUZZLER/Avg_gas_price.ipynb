```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
import pandas as pd

# Read data from text file
with open("avg_gasprice.txt", "r") as f:
    data = f.read()

# Clean up data and split into list of lists
data = data.replace("[[", "").replace("]]", "").replace("], [", "|")
data = data.split("|")
data = [d.split(", ") for d in data]

# Create DataFrame from list of lists
df = pd.DataFrame(data, columns=["Month/Year", "Average_gas_price"])

# Convert Average_gas_used column to float type
df['Average_gas_price'] = df['Average_gas_price'].astype('float')

#Convert the Month/Year column to datetime datatype and sorting them in ascending order.
df['Month/Year']=df['Month/Year'].apply(lambda x: x.strip('"'))
df['Month/Year'] = pd.to_datetime(df['Month/Year'], format='%m/%Y')
df = df.sort_values(by='Month/Year')
# Print the DataFrame
df
```

```python
plt.figure(figsize = (15,10))
plt.title("Average gas price change over time")
plt.xticks(rotation=90)
plt.plot(df['Month/Year'], df['Average_gas_price'])
plt.xlabel("Year-Month")
plt.ylabel("Average gas price")
plt.savefig("Average_gas_price.png")  
plt.show()
```
