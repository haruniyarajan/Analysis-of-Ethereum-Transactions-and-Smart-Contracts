The average value of transaction in each month between the start and end of the dataset.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
import pandas as pd

with open('transactions_avg.txt', 'r') as file:
    data = file.read().splitlines()

    # remove the square brackets and quotation marks from the string
data_str = data[0].replace('[','').replace(']','').replace('"','')

# split the string into a list of values
data_list = data_str.split(',')

# create a dictionary from the list
data_dict = {'Month/Year':[], 'Average_transaction':[]}
for i in range(0,len(data_list),2):
    data_dict['Month/Year'].append(data_list[i])
    data_dict['Average_transaction'].append(data_list[i+1])

# create the DataFrame
df = pd.DataFrame(data_dict)

# Convert transaction column to float type
df['Average_transaction'] = df['Average_transaction'].astype("float")

# print the DataFrame
df
```

```python
plt.figure(figsize = (15,10))
plt.title("Average transactions occurring every month")
plt.xticks(rotation=90)
plt.bar(df['Month/Year'], df['Average_transaction'])
plt.xlabel("Year-Month")
plt.ylabel("Average transaction")
plt.savefig("Avg_trans_per_month")
plt.show()
```
