 A bar plot to show the number of transactions occurring every month between the start and end of the dataset.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
import pandas as pd

# Read data from text file
with open("no_of_transactions.txt", "r") as f:
    data = f.read()

# Clean up data and split into list of lists
data = data.replace("[[", "").replace("]]", "").replace("], [", "|")
data = data.split("|")
data = [d.split(", ") for d in data]

# Create DataFrame from list of lists
df = pd.DataFrame(data, columns=["Year/Month", "Transactions"])

# Convert transaction column to numeric type
df['Transactions'] = pd.to_numeric(df['Transactions'])

# Print the DataFrame
df
```

```python
plt.figure(figsize = (15,10))
plt.title("Number of transactions occurring every month")
plt.xticks(rotation=90)
plt.bar(df['Year/Month'], df['Transactions'])
plt.xlabel("Year-Month")
plt.ylabel("Number of transactions")
plt.savefig("No_of_trans_per_month.png")  
plt.show()
```
